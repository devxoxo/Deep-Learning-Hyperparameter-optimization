```python
from tensorflow.keras.datasets import mnist

(X_train, Y_train), (X_test, Y_test) = mnist.load_data()
X_test
X_train = X_train.reshape(60000, 28,28,1)
X_test = X_test.reshape(10000, 28,28,1)
X_train = X_train/255.0
X_test = X_test/255.0

Y_train = to_categorical(Y_train)
Y_test = to_categorical(Y_test)
```

```python
tf.debugging.set_log_device_placement(True)
```

```python

#strategy =  tf.distribute.MirroredStrategy()
strategy = tf.distribute.MirroredStrategy(devices=["/gpu:0", "/gpu:1"],cross_device_ops=tf.distribute.HierarchicalCopyAllReduce())

print('Number of devices: {}'.format(strategy.num_replicas_in_sync))
with strategy.scope():
    model = Sequential()
    # Block 1
    model.add(Conv2D(filters=64, kernel_size=(3,3), padding='same', strides=1, activation='relu', input_shape=(28,28,1,), name='block1_conv1'))
    model.add(Conv2D(filters=64, kernel_size=(3,3), padding='same', strides=1, activation='relu',  name='block1_conv2'))
    model.add(MaxPool2D(pool_size=(2,2), strides=1, name='block1_pool'))

    # Block 2
    model.add(Conv2D(filters=128, kernel_size=(3,3), padding='same', strides=1, activation='relu',  name='block2_conv1'))
    model.add(Conv2D(filters=128, kernel_size=(3,3), padding='same', strides=1, activation='relu',  name='block2_conv2'))
    model.add(MaxPool2D(pool_size=(2,2), strides=2,name='block2_pool'))

    # Block 3
    model.add(Conv2D(filters=256, kernel_size=(3,3), padding='same', strides=1, activation='relu',  name='block3_conv1'))
    model.add(Conv2D(filters=256, kernel_size=(3,3), padding='same', strides=1, activation='relu',  name='block3_conv2'))
    model.add(Conv2D(filters=256, kernel_size=(3,3), padding='same', strides=1, activation='relu',  name='block3_conv3'))
    model.add(Conv2D(filters=256, kernel_size=(3,3), padding='same', strides=1, activation='relu',  name='block3_conv4'))
    model.add(MaxPool2D(pool_size=(2,2), strides=2,name='block3_pool'))

    # Block 4
    model.add(Conv2D(filters=512, kernel_size=(3,3), padding='same', strides=1, activation='relu',  name='block4_conv1'))
    model.add(Conv2D(filters=512, kernel_size=(3,3), padding='same', strides=1, activation='relu',  name='block4_conv2'))
    model.add(Conv2D(filters=512, kernel_size=(3,3), padding='same', strides=1, activation='relu',  name='block4_conv3'))
    model.add(Conv2D(filters=512, kernel_size=(3,3), padding='same', strides=1, activation='relu',  name='block4_conv4'))
    model.add(MaxPool2D(pool_size=(2,2), strides=2,name='block4_pool'))

    # Block 5
    model.add(Conv2D(filters=512, kernel_size=(3,3), padding='same', strides=1, activation='relu',  name='block5_conv1'))
    model.add(Conv2D(filters=512, kernel_size=(3,3), padding='same', strides=1, activation='relu',  name='block5_conv2'))
    model.add(Conv2D(filters=512, kernel_size=(3,3), padding='same', strides=1, activation='relu',  name='block5_conv3'))
    model.add(Conv2D(filters=512, kernel_size=(3,3), padding='same', strides=1, activation='relu',  name='block5_conv4'))
    model.add(MaxPool2D(pool_size=(2,2), strides=2,name='block5_pool'))



    model.add(Flatten())
    model.add(Dense(1024, activation='relu', name='fc1'))
    model.add(Dropout(0.2))
    model.add(BatchNormalization())
    model.add(Dense(128, activation='relu', name='fc2'))
    model.add(BatchNormalization())
    model.add(Dense(10, activation='softmax', name='predictions')) #최종은 항상 클래스 종류의 개수 

    model.compile(loss='categorical_crossentropy', optimizer=optimizers.Adam(lr = 0.00001), metrics=['accuracy'])

    



    
    
model.summary()
```

```python
hist = model.fit(X_train, Y_train, batch_size=500, epochs=4, validation_split=0.2)
```

```python
score = model.evaluate(X_test, Y_test)
score
```

```python
plt.figure(figsize=(10,10))
plt.subplot(2, 2, 1)
plt.plot(hist.history['accuracy'], color='r')
plt.plot(hist.history['val_accuracy'], color='b')
plt.title('accuracy')

#loss 조회
plt.subplot(2, 2, 2)
plt.plot(hist.history['loss'], color='r')
plt.plot(hist.history['val_loss'], color='b')
plt.title('loss')
plt.show()
```
